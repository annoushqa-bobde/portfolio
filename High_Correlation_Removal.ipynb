```python
import numpy as np
import pandas as pd
```

```python
def remove_high_correlation_predictors(data, threshold=0.75):
    corr_matrix = data.corr().abs()
    initial_columns = len(data.columns)
    print("Initial number of columns:", initial_columns)

    np.fill_diagonal(corr_matrix.values, 0)

    while True:
        max_corr = np.unravel_index(np.argmax(corr_matrix), corr_matrix.shape)
        max_corr_value = corr_matrix.iloc[max_corr]

        if max_corr_value <= threshold:
            break

        avg_corr_A = corr_matrix.iloc[max_corr[0]].mean()
        avg_corr_B = corr_matrix.iloc[max_corr[1]].mean()

        if avg_corr_A > avg_corr_B:
            removed_predictor = data.columns[max_corr[1]]
            data = data.drop(removed_predictor, axis=1)
        else:
            removed_predictor = data.columns[max_corr[0]]
            data = data.drop(removed_predictor, axis=1)

        corr_matrix = data.corr().abs()
        np.fill_diagonal(corr_matrix.values, 0)

    final_columns = len(data.columns)
    print("Final number of columns:", final_columns)

    return data
```

```python
data=pd.read_csv('/segmentationOriginal.csv')
del data['Unnamed: 0']
data.groupby('Case')['Case'].count()
cell_train= data[data['Case']=='Train']
cell_test= data[data['Case']=='Test']
cell_train_feature= cell_train.iloc[:,3:]
cell_train_feature.head()
```

```python
data2= remove_high_correlation_predictors(cell_train_feature, threshold=0.75)
predictors= list(data2.columns)
corr_matrix= np.corrcoef(data2, rowvar=False)
print(corr_matrix)

threshold=0.75
max_corr= np.max(np.abs(corr_matrix-np.eye(data2.shape[1])))
if max_corr<threshold:
  print("All pairwise correlation is below the threshold")
else:
  print("All pairwise correlation is not below the threshold")
```
